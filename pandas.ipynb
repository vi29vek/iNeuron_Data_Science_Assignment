```python
import pandas as pd
# In pandas, there are mainly two datatypes
# one is dataframe which is equivalent to tables in sql
# the other one is series, which is equivakent to a python list
# the columns and rows combines are called dataframe
# the columns are of type series
# Dataframe is a kind of dictionary, in which all the columns are keys,and the series is the value of each key
```

```python
# To load an excel file we use pd.read_excel
# it returns a dataframe which is similar to table in sql
excel_df = pd.read_excel(r'C:\MyData\MyData\data science\resources\data sets\Attribute DataSet.xlsx')
```

```python
excel_df
```

```python
# We can have multiple sheets in out excel. 
# By default read_excel reads the first sheet
# In order to read from some other sheet of the same excel file-
pd.read_excel(r'C:\MyData\MyData\data science\resources\data sets\Attribute DataSet.xlsx',sheet_name='Sheet2')
```

```python
# If we dont want our headers to be a header and want them to be a record, we set header=None while reading excel file
pd.read_excel(r'C:\MyData\MyData\data science\resources\data sets\Attribute DataSet.xlsx',header=None)
```

```python
# If we want, we can give column name by our self as well-
pd.read_excel(r'C:\MyData\MyData\data science\resources\data sets\Attribute DataSet.xlsx',header=None,sheet_name='Sheet2',names=['first','second','third','fourth'])
```
