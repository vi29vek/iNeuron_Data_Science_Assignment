# How to drop a column

```python
# axis tells wether to delete a column or a row
# axis = 1 or 'columns' deletes the column
# axis = 0 or 'index' deletes the column
# By default, axis is set to 0
df.drop('cost_to_company',axis=1)
```

```python
# This drop method doesnt deletes column from actual dataframe.
# If we check, our dataframe is still same
df
```

```python
# If we want drop method to change the dataframe, we have to set the attribute 'inplace' to True
df.drop('cost_to_company',axis=1,inplace=True)
```

```python
df
```
